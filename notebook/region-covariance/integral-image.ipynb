# Integral Image

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import seaborn as sns
```

```python
img = np.arange(1, 10, dtype=np.uint8).reshape(3, 3)

plt.rcParams["font.size"] = 30
sns.heatmap(img, vmin=0, vmax=np.sum(img), square=True, annot=True, cbar=False, xticklabels=False, yticklabels=False, fmt='d')
plt.grid(False)
plt.rcParams["font.size"] = 10
plt.title('Original Image')
plt.show()
```

```python
plt.rcParams["font.size"] = 30
sns.heatmap(cv2.integral(img), vmin=0, vmax=np.sum(img), square=True, annot=True, cbar=False, xticklabels=False, yticklabels=False, fmt='d')
plt.grid(False)
plt.rcParams["font.size"] = 10
plt.title('Integral Image')
plt.show()
```
